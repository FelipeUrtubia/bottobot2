```python
tamaño = int(input('ingrese el tamaño del largo del cuadro'))
pm = int(input("ingrese el porcentaje de mutacion"))
ng = int(input("ingrese el numero de generaciones"))
pob = int(input("ingrese el tanaño de la poblacion"))
url = input("ingrese la url del github")
print("=====================================================================================================================================")
print("tipo de cuadro: " + str(tamaño) + "x10 " +" poblacion: " + str(pob) + " numero de generaciones: " + str(ng) + " mutacion: " + str(pm) + " url: " + str(url))
print("=====================================================================================================================================")
```

```python
from urllib.request import urlopen


txt = urlopen(url).read().decode("utf-8") 
  
txt = txt.split("\n")
txt.pop(10)
for line in txt:
    print(line)
```

```python
resto = 0
gen =  []
for line in txt:
  if len(line) < 10:
    resto = 10-len(line)
    for i in line:
      if i == " ":
        gen.append("null")
      elif i == "X":
        gen.append(0)
      elif i == "P":
        gen.append("P")
      elif i == "M":
        gen.append("M")
    for i in range(resto):
      gen.append("null")
  else:
    for i in line:
      if i == " ":
        gen.append("null")
      elif i == "X":
        gen.append(0)
      elif i == "P":
        gen.append("P")
      elif i == "M":
        gen.append("M")
print(gen)
```

```python
#UNA EXPLICACION DE LA PARAMETRIZACION EN:
#https://blog.paperspace.com/working-with-different-genetic-algorithm-representations-python/
import pygad
import numpy


#Genera un cromosoma de puros 1s
num_generations = int(ng) # Number of generations.
num_parents_mating = 7 # Number of solutions to be selected as parents in the mating pool.
sol_per_pop = int(pob) # Number of solutions in the population.
num_genes = tamaño*10
parent_selection_type = "tournament"  # Type of parent selection.
keep_parents = 7 # Number of parents to keep in the next population. -1 means keep all parents and 0 means keep nothing.
crossover_type = "single_point" # Type of the crossover operator.
# Parameters of the mutation operation.
mutation_type = "random" # Type of the mutation operator.
mutation_percent_genes = int(pm) # Percentage of genes to mutate. This parameter has no action if the parameter mutation_num_genes exists or when mutation_type is None.


def fitness_func(solution, parametro2):
  comparation = solution
  fitness = 0
  for i in range(num_genes):
    if comparation[i] == 1 and  gen[i]== 0:
       if i == 0 :
         fitness = fitness+1
    
       elif i > 0 and i <= 9:
         if comparation[i-1] == 0:
           fitness = fitness+1

       elif i > 9 and str(i)[-1] == 9:
          if comparation[i-10] == 0 and comparation[i-11] == 0:
            fitness = fitness+1

       elif i > 1 and str(i)[-1] == 0 and i:
          if comparation[i-10] == 0 and comparation[i-9] == 0:
            fitness = fitness+1

       else:
          if comparation[i-11] == 0 and comparation[i-10] == 0 and comparation[i-10] == 0 and comparation[i-9] == 0 and comparation[i-1] == 0:
            fitness = fitness+1    
          
    if gen[i] == "P" and comparation[i] == 1:
      fitness = fitness-1
    if gen[i] == "null" and comparation[i] == 1:
      fitness = fitness-1
    if gen[i] == "M":
      if comparation[i-1] == 1:
        if comparation[i-10] == 1 or comparation[i+1] ==1 or comparation[i+10] == 1:
          fitness = fitness-1
      if comparation[i-10] == 1:
         if comparation[i-1] == 1 or comparation[i+1] ==1 or comparation[i+10] == 1:
          fitness = fitness-1
      if comparation[i+1] == 1:
         if comparation[i-1] == 1 or comparation[i+10] ==1 or comparation[i-10] == 1:
          fitness = fitness-1
      if comparation[i+10] == 1:
         if comparation[i-1] == 1 or comparation[i+1] ==1 or comparation[i-10] == 1:
          fitness = fitness-1

  return fitness

fitness_function = fitness_func


last_fitness = 0
bf = -999
gc = gen
def callback_generation(ga_instance):
    global last_fitness
    global bf
    global bs
    global gc
    print("Generation = {generation}".format(generation=ga_instance.generations_completed),
    ",   Fitness Best   = {fitness:.3f}".format(fitness=ga_instance.best_solution()[1]),
    ",   Change     = {change:.3f}".format(change=ga_instance.best_solution()[1] - last_fitness))
    last_fitness = ga_instance.best_solution()[1]
    solution, solution_fitness, solution_idx = ga_instance.best_solution()
    ps = []
    loc = 0
    sol = solution
    aline = ""
    for i in range(len(sol)):
      ps.append(int(sol[i]))
    i = 0
    while i < len(sol):
      aline  = ""
      for j in range(10):
        cc = str(int(sol[loc]))
        if gc[loc] == "null":
          cc = " "
          sol[loc] = 0
        elif gc[loc] == "M":
          cc = "M"
          sol[loc] = 0
        elif gc[loc] == "P":
          cc = "P"
          sol[loc] = 0
        elif gc[loc] == 0 and cc == "1":
          if loc == 0:
            cc = "X"
          if loc > 0 and loc <= 9:
            if sol[loc - 1 ] == 1:
              sol[loc] = 0
              cc = "0"
            else:
             cc = "X"  
          if loc > 9 and loc <  len(sol)- 9 and str(loc)[-1] == 0:
            if  sol[loc - 9] == 1 or sol[loc -10] == 1:
              sol[loc] = 0
              cc = "0"
            else:
              cc = "X"
          if loc > 9  and str(loc)[-1] == 9: 
            if sol[loc -1] == 1 or sol[loc -10] == 1 or sol[loc -9]  == 1:
              sol[loc] = 0
              cc = "0"
            else:
              cc = "X"
          else:
            if sol[loc -1] == 1 or sol[loc - 9] == 1 or sol[loc - 10] == 1 or sol[ loc -11] == 1:
              cc = "0"
              sol[loc] = 0
            else:
              cc = "X"
   
        loc = loc + 1
        aline = aline + " " + cc
        i += 1
      print(aline)
    if solution_fitness > bf :
      bf = solution_fitness
      bs = solution
 
ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       sol_per_pop=sol_per_pop,
                       num_genes=num_genes,
                       fitness_func=fitness_function,
                       parent_selection_type=parent_selection_type,
                       gene_space=[0, 1],
                       keep_parents=keep_parents,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes,
                       callback_generation=callback_generation)

# Running the GA to optimize the parameters of the function.
ga_instance.run()

# Returning the details of the best solution.
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("la mejor generacion es la numero {best_solution_generation} .".format(best_solution_generation=ga_instance.best_solution_generation))
print("su fitness es :" + str(bf))
print("sus parametros son: ")
ps = []
loc = 0
aaline = ""
for i in range(len(bs)): 

  ps.append(int(bs[i]))
i = 0
while i < len(bs):
  aline  = ""
  j = 0
  for j in range(10):
    cc = str(int(bs[loc]))
    if gc[loc] == "null":
      cc = " "
      bs[loc] = 0
    elif gc[loc] == "M":
      cc = "M"
      bs[loc] = 0
    elif gc[loc] == "P":
      cc = "P"
      bs[loc] = 0
    elif gc[loc] == 0 and cc == "1":
      if loc == 0:
        cc = "X"
        bs[loc] = 0
      if loc > 0 and loc <= 9:
        if bs[loc - 1 ] == 1:
          bs[loc] = 0
          cc = "0"
        else:
          cc = "X"  
      if loc > 9 and loc <  len(bs)- 9 and str(loc)[-1] == 0:
        if  bs[loc - 9] == 1 or bs[loc -10] == 1:
          bs[loc] = 0
          cc = "0"
        else:
          cc = "X"
      if loc > 9  and str(loc)[-1] == 9: 
        if bs[loc -1] == 1 or bs[loc -10] == 1 or bs[loc -9]  == 1:
          bs[loc] = 0
          cc = "0"
        else:
          cc = "X"
      else:
        if bs[loc -1] == 1 or bs[loc - 9] == 1 or bs[loc - 10] == 1 or bs[ loc -11] == 1:
          cc = "0"
          bs[loc] = 0
        else:
          cc = "X"
   
    loc = loc + 1
    aline = aline + " " + cc
  print(aline)
  i = i + 10
          
# Saving the GA instance.
filename = 'genetic' # The filename to which the instance is saved. The name is without extension.
ga_instance.save(filename=filename)

# Loading the saved GA instance.
loaded_ga_instance = pygad.load(filename=filename)
loaded_ga_instance.plot_result()
```

```python
#UNA EXPLICACION DE LA PARAMETRIZACION EN:
#https://blog.paperspace.com/working-with-different-genetic-algorithm-representations-python/

import pygad
import numpy

#CALCULA EL VALOR DE X EN LA ECUASION,    0 = 55 - 12.3*pow(X, 3), USANDO AG BINARIO

num_generations = 100 # Number of generations.
num_parents_mating = 7 # Number of solutions to be selected as parents in the mating pool.
sol_per_pop = 20 # Number of solutions in the population.
num_genes = 20
parent_selection_type = "tournament"  # Type of parent selection.
keep_parents = 7 # Number of parents to keep in the next population. -1 means keep all parents and 0 means keep nothing.
crossover_type = "single_point" # Type of the crossover operator.
# Parameters of the mutation operation.
mutation_type = "random" # Type of the mutation operator.
mutation_percent_genes = 10 # Percentage of genes to mutate. This parameter has no action if the parameter mutation_num_genes exists or when mutation_type is None.


def fitness_func(solution, parametro2):
    #Convierte el cromosoma de 1s y 0s en un decimal
    decimal = int(''.join(map(lambda solution: str(int(solution)), solution)), 2)  / 1000
    fitness = 1.0/abs(55 - 12.3*pow(decimal, 3))
    #print(decimal, - 12.3*pow(decimal, 3),  fitness)
    return fitness

fitness_function = fitness_func


last_fitness = 0
def callback_generation(ga_instance):
    global last_fitness
    print("Generation = {generation}".format(generation=ga_instance.generations_completed),
    ",   Fitness Best   = {fitness:.3f}".format(fitness=ga_instance.best_solution()[1]),
    ",   Change     = {change:.3f}".format(change=ga_instance.best_solution()[1] - last_fitness))
    last_fitness = ga_instance.best_solution()[1]


ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       sol_per_pop=sol_per_pop,
                       num_genes=num_genes,
                       fitness_func=fitness_function,
                       parent_selection_type=parent_selection_type,
                       gene_space=[0, 1],
                       keep_parents=keep_parents,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes,
                       callback_generation=callback_generation)

# Running the GA to optimize the parameters of the function.
ga_instance.run()

# Returning the details of the best solution.
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Parameters of the best solution : {solution}".format(solution=solution))
print("Fitness value of the best solution = {solution_fitness}".format(solution_fitness=solution_fitness))
decimal = int(''.join(map(lambda solution: str(int(solution)), solution)), 2)  / 1000
print("Fitness value of - 12.3*pow(decimal, 3) = ",  - 12.3*pow(decimal, 3))
print("Index of the best solution : {solution_idx}".format(solution_idx=solution_idx))


if ga_instance.best_solution_generation != -1:
    print("Best fitness value reached after {best_solution_generation} generations.".format(best_solution_generation=ga_instance.best_solution_generation))

# Saving the GA instance.
filename = 'genetic' # The filename to which the instance is saved. The name is without extension.
ga_instance.save(filename=filename)

# Loading the saved GA instance.
loaded_ga_instance = pygad.load(filename=filename)
loaded_ga_instance.plot_result()
```
