```python
room = ['x  xxxxxxP', 'x  xxMxxxx', 'x  xxxxxxx', 'xxxxxxxxxx', 'xMxxxxxxxx', 'xxxxxxxxxP', 'xxxxx     ', 'xxMxx     ', 'xxxxx     ', 'xxxxx     ']

print(len(room))
print(room)

for line in room:
    print(line)
    
resto = 0
gen =  []
for line in room:
  if len(line) < 10:
    resto = 10-len(line)
    for i in line:
      if i == " ":
        gen.append("null")
      elif i == "x":
        gen.append(0)
      elif i == "P":
        gen.append("P")
      elif i == "M":
        gen.append("M")
    for i in range(resto):
      gen.append("null")
  else:
    for i in line:
      if i == " ":
        gen.append("null")
      elif i == "x":
        gen.append(0)
      elif i == "P":
        gen.append("P")
      elif i == "M":
        gen.append("M")
print(gen)
```
